## Preprocessing of Data

### ID Column

```python
df= df.drop('ID',axis=1)
```

```python
df.head()
```

### Reason Column

```python
df['Reason for Absence'].nunique()
```

```python
len(df['Reason for Absence'].unique())
```

```python
df['Reason for Absence'].sort_values().unique()
```

```python
reason_column= pd.get_dummies(df['Reason for Absence'],drop_first=True)
```

```python
df= df.drop('Reason for Absence',axis=1)
```

```python
df.head()
```

```python
reason_1=reason_column.loc[:,1:14].max(axis=1)
reason_2=reason_column.loc[:,15:17].max(axis=1)
reason_3=reason_column.loc[:,18:21].max(axis=1)
reason_4=reason_column.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df,reason_1,reason_2,reason_3,reason_4],axis=1)
df.head()
```

```python
df.columns.values
```

```python
column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns=column_names
```

```python
df.head()
```

```python
column_reorder=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours' ]
```

```python
df = df[column_reorder]
```

```python
df.head()
```

### Date Column

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
df['month']=df.Date.dt.month
```

```python
df['Day of year']=df.Date.dt.dayofyear
```

```python
df['Day of week']=df.Date.dt.dayofweek
```

```python
df.head(8)
```

```python
df=df.drop('Date',axis=1)
```

### Education

```python
df['Education'].nunique()
```

```python
df['Education'].unique()
```

```python
df['Education'].value_counts()
```

**So we group 1 as 0 and 2,3,4 as 1**

```python
df['Education']= df['Education'].map({1:0,2:1,3:1,4:1})
```
